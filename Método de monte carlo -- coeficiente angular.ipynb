Suponha que você quer analisar o que ocorre com a distribuição do estimador de mínimos quadrados do coeficiente angular da reta da forma y = α + βx + u quando aumentamos o tamanho da amostra. Suponha que u é uma variável aleatória normal padrão e α = 2 e β = 3. Vamos fazer o papel de Deus e gerar os valores de y usando os vetores [x, y] para estimar os coeficientes α e β (embora estejamos aqui particularmente interessados nas estimativas de β).

```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Função para gerar os dados
def geracaoDados(n):
    alpha = 2
    beta = 3
    x = np.random.normal(0, 4, n)  # Gerando valores de x a partir de uma distribuição normal
    u = np.random.normal(0, 1, n)  # Gerando valores de u a partir de uma distribuição normal. u normal ~(0,1)
    y = alpha * np.ones(n) + beta * x + u  # Calculando y = alpha + beta * x + u
    return x, y

# Função para estimar os coeficientes
def estimacao(x, y):
    beta, alpha, r_value, p_value, std_err = stats.linregress(x, y)  # Realizando a regressão linear
    return alpha, beta

def plot_histogram(n, title):
    # Cria um vetor vazio para armazenar os coeficientes beta
    vetor = np.empty(numeroAmostras)
    
    # Loop para gerar as amostras e estimar os coeficientes beta
    for i in range(numeroAmostras):
        x, y = geracaoDados(n)
        alpha, beta = estimacao(x, y)
        vetor[i] = beta
    
    # Plota o histograma dos coeficientes beta
    plt.hist(vetor, bins=30, edgecolor='black')
    plt.axvline(np.mean(vetor), color='r', linestyle='--', linewidth=2)
    plt.title(title)
    plt.show()

if __name__ == '__main__':
    numeroAmostras = 1000

    sizes = [10, 1000]
    titles = ['Amostra de tamanho 10', 'Amostra de tamanho 1000']

    # Loop para gerar e plotar os histogramas para cada tamanho de amostra
    for n, title in zip(sizes, titles):
        plot_histogram(n, title)
```

O exercício proposto envolve analisar a distribuição do estimador de mínimos quadrados do coeficiente angular de uma reta conforme aumentamos o tamanho da amostra.

Ao gerar dados fictícios e estimar os coeficientes da reta usando o método dos mínimos quadrados, obtemos diferentes valores para o coeficiente angular. Repetindo esse processo várias vezes e armazenando os valores do coeficiente angular em um vetor, podemos visualizar a distribuição desses valores por meio de um histograma.

Espera-se que, ao aumentar o tamanho da amostra, a distribuição do estimador de mínimos quadrados seja mais centrada no verdadeiro valor do coeficiente angular. Isso ocorre porque um tamanho de amostra maior fornece mais informações para a estimativa, reduzindo a incerteza e a variabilidade do estimador.

Ao comparar os histogramas das amostras de tamanhos 10 e 1000, podemos observar como a distribuição do estimador de mínimos quadrados muda à medida que o tamanho da amostra aumenta. Essa análise nos ajuda a entender a precisão e confiabilidade da estimativa do coeficiente angular em diferentes cenários de tamanho amostral.

Ao aumentar o tamanho da amostra, espera-se que a variância do estimador de mínimos quadrados do coeficiente angular seja reduzida. Isso significa que os valores estimados do coeficiente angular terão menos dispersão em torno do valor verdadeiro.

Dessa forma, o aumento do tamanho da amostra não apenas leva a uma estimativa mais próxima do valor verdadeiro do coeficiente angular, mas também reduz a variabilidade dos valores estimados. Isso resulta em estimativas mais precisas e consistentes do coeficiente angular.
